# LOC(Location)

```python
# Change the values using LOC function 

dataframe.loc[0, 0] = 3000
dataframe.loc[0, 1] = 4000
dataframe.loc[0, 2] = 5000
```

```python
dataframe.head()
```

```python
# Change the column index 
dataframe.columns = list("ABCDE")
```

```python
dataframe.head()
```

```python
# adding a new column

# 0th row & 0th column
dataframe.loc[0, 0] = 10
```

```python
dataframe.head()
```

```python
# addding a new row

# 'a'th row & 0th column
dataframe.loc['a',0]=2121
```

```python
dataframe.tail()
```

```python
# Deleting a row or column

# 0th from the row
dataframe.drop(0, axis=0)
```

```python
# Deleting a row or column: just returns a copy of deleted data

# 0th from the column
dataframe.drop(0, axis=1)
```

```python
 # Set of a Big dataframe

# returns a sub dataframe of rows (1,2) & columns('C', 'D')
newdf = dataframe.loc[[1,2], ['C', 'D']]
```

```python
newdf
```

```python
# All rows & all columns

# all column with rows 1&2
newdf = dataframe.loc[[1,2], : ]
```

```python
newdf
```

```python
# all rows with 'C', 'D' columns
newdf = dataframe.loc[:, ['C', 'D']]
```

```python
newdf
```

```python
# whole table
newOne = dataframe.loc[:]
```

```python
newOne
```

```python
# Database queries

# prints all the table where data in 'A' th column < 0.3
dataframe.loc[ (dataframe['A'] < 0.3) ]
```

```python
# prints all the table where data in 'A' th column < 0.3 && 'C' th column > 0.1
dataframe.loc[ (dataframe['A'] < 0.3) & (dataframe['C'] > 0.1) ]
```

# iLoc(indexed location)

```python
# get the values using proper index for rows & columns

# 0th row & 4th columns
dataframe.iloc[0,4]
```

```python
# multiple rows & columns
dataframe.iloc[[1,3], [3,2]]
```

```python
# Deleting & updating the original dataframe as the same time

# deletes 3rd named row & updated original dataframe
dataframe.drop([3], inplace=True)
```

```python
# 3 named row is deleted
dataframe.head(5)
```

```python
# Reset the index

# resets the index, with extra index column
dataframe.reset_index()
```

```python
# without extra index column
dataframe.reset_index(drop=True)
```

```python
# modify original dataframe inplace
dataframe.reset_index(drop=True, inplace=True)
```
