```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# np.all() treats every element in it as a boolean & eveluates it as true 
# until it is not explicitly 'false' '0' 'None'

print(np.all([[1, 1],[1, 0]]))
print(np.all([-1, 4, 5]))
```

```python
# creating x and y axis to plot

# X axis
x = [1,2,3,4]

# Y axis
y = [12,24,36,48]
```

```python
# plotting the values using pyplot
plt.plot(x,y)
```

```python
# to show on the screen
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])  

# setting x axis limit to (0,6) and y axis to (0,20)
plt.axis([0, 6, 0, 20])  

plt.show()

```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])  

# off the x and y axis 
plt.axis('off')
```
